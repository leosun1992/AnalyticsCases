```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("ML-EMHYY.csv")
```

```python
df.head()
```

```python
def arrange_time_dataframe(df):
    df.columns = ['date', 'value']
    df.sort_values(by='date', inplace=True)
    df.set_index('date', inplace=True)
    return df
```

```python
df = arrange_time_dataframe(df)
```

```python
df.head()
```

```python
df.plot(figsize=(16, 6))
```
